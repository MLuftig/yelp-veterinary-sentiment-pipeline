```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
tf_idf = pd.read_csv('/kaggle/input/datasets/micahluftig/yelp-veterinary-reviews-tfidf/tfidf_diff_top_words.csv')
```

```python
print(tf_idf.head())
```

```python
# sort so most "low rating" words are at the top, most "high rating" at the bottom
plot_df = tf_idf.sort_values('diff', ascending=False).reset_index(drop=True)

colors = ['#d62728' if v > 0 else '#1f77b4' for v in plot_df['diff']]

fig, ax = plt.subplots(figsize=(8, 10))
ax.barh(plot_df['word'], plot_df['diff'], color=colors)

ax.axvline(0, color='black', linewidth=0.8)
ax.set_xlabel('← associated with high ratings        associated with low ratings →')
ax.set_title('Words that distinguish high-star from low-star vet reviews', fontsize=13, pad=15)

# invert so the first row (highest positive diff) is at the top
ax.invert_yaxis()

plt.tight_layout()
plt.savefig('/kaggle/working/tfidf_diverging_chart.png', dpi=200, bbox_inches='tight')
plt.show()
```
